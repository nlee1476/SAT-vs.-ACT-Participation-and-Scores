#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
def std_dev_calculator(series):
    n = len(series)
    mu = np.mean(series)
    count = 0
    for i in series: 
        count += (i - mu)**2
    if i == str:
        pass
        
    count_over_n = count/n
    sigma = (count_over_n)**0.5
    return sigma
````

````python
std_dev_calculator(sat_2017["sat_math_2017"])
````
